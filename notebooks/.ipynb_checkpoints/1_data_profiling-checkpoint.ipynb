```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
pd.set_option('display.precision', 3)
```

```python
DATA_PATH = Path("../data/raw")

tables = {}
for file in DATA_PATH.glob('*.csv'):
    tables[file.stem] = pd.read_csv(file)

tables.keys()
```

```python
sellers_df = tables['olist_sellers_dataset']
product_category_df = tables['product_category_name_translation']
orders_df = tables['olist_orders_dataset']
order_items_df = tables['olist_order_items_dataset']
customers_df = tables['olist_customers_dataset']
geolocation_df = tables['olist_geolocation_dataset']
order_payments_df = tables['olist_order_payments_dataset']
order_reviews_df = tables['olist_order_reviews_dataset']
products_df = tables['olist_products_dataset']
```

```python
def dataset_info(df, name):
    print('-' * 140)
    print(name)
    print(f'Shape: {df.shape}')
    display(df.head())
    display(df.info(memory_usage='deep'))
    display(df.describe(include="all").T)

for name, df in tables.items():
    dataset_info(df, name)
```

```python
def missing_info(df, name):
    print(f'Пропущенные значения в таблице {name}: ')
    display(df.isnull().sum())
    print()

for name, df in tables.items():
    missing_info(df, name)
```

```python
print('Количество дубликатов в каждой таблице:')
for name, df in tables.items():
    print(f'{name}: {df.duplicated().sum()}')
```

```python
print('Проверка уникальности первичного ключа в каждой таблице:')
print(f'sellers_df: {sellers_df.duplicated(['seller_id']).sum()}')
print(f'product_category_df: {product_category_df.duplicated(['product_category_name']).sum()}')
print(f'orders_df: {orders_df.duplicated(['order_id']).sum()}')
print(f'order_items_df: {order_items_df.duplicated(subset=['order_id', 'order_item_id']).sum()}')
print(f'customers_df: {customers_df.duplicated(['customer_id']).sum()}')
print(f'order_payments_df: {order_payments_df.duplicated(subset=['order_id', 'payment_sequential']).sum()}') 
print(f'order_reviews_df: {order_reviews_df.duplicated(subset=['review_id', 'order_id']).sum()}')
print(f'products_df: {products_df.duplicated(['product_id']).sum()}')
print('\nПроверка уникальности комбинации координат и ZIP-кода, т.к. PK отсутствует:')
print(f'geolocation_df: {geolocation_df.duplicated(['geolocation_zip_code_prefix', 'geolocation_lat', 'geolocation_lng']).sum()}')
```

```python
print('Типы данных каждой таблицы:')
for name, df in tables.items():
    print(f'\n{name}:')
    display(df.dtypes)
```

---------------------------------------------------------------------------------------------------------------------------------------------
Всего исследовано 9 таблиц.

Основные выводы:
- Полные дубликаты обнаружены только в geolocation и обусловлены логикой хранения координат по ZIP-коду
- Нарушений уникальности первичных ключей не найдено (кроме geolocation, где PK отсутствует)
- Основные пропуски сосредоточены в orders, reviews и products. Пропуски в orders обусловлены в основном незавершённостью или отменённостью заказов, пропуски в reviews из-за отсутствия текстовых отзывах, что является нормой, в products - частичные пропуски описания товаров
- Все столбцы с датами загружены как object и требуют преобразования в datetime
