### Проверка пропущенных значений
---------------------------------------------------------------------------------------------------------------------------------------------

```python
def missing_info(df, name):
    print(f'Пропущенные значения в таблице {name}: ')
    df_missing = df.isnull().sum().to_frame("missing_count")
    df_missing['missing_percent'] = (df_missing['missing_count'] / len(df) * 100).round(2)
    display(df_missing)
    print()

for name, df in tables.items():
    missing_info(df, name)
```

### Проверка логики дат
---------------------------------------------------------------------------------------------------------------------------------------------

```python
reviews_date_df = order_reviews_df[['order_id','review_creation_date','review_answer_timestamp']]
merge_orders_reviews = orders_df.merge(reviews_date_df, how='left', on='order_id')
```

```python
def check_anomaly(df, condition, message):
    mask = df.eval(condition)
    count = mask.sum()
    print(f"{message}: {count}")
    if count > 0:
        display(df[mask].head())

check_anomaly(
    orders_df,
    'order_delivered_customer_date < order_purchase_timestamp',
    'Доставка раньше оформления'
)

check_anomaly(
    orders_df,
    'order_approved_at < order_purchase_timestamp',
    'Подтверждение раньше оформления'
)

check_anomaly(
    orders_df,
    'order_delivered_customer_date < order_delivered_carrier_date',
    'Передача в доставку позже доставки заказа'
)

check_anomaly(
    orders_df,
    'order_delivered_carrier_date < order_purchase_timestamp',
    'Передача в доставку раньше оформления'
)

check_anomaly(
    order_reviews_df,
    'review_answer_timestamp < review_creation_date',
    'Ответ на отзыв раньше, чем отзыв был оставлен'
)

check_anomaly(
    merge_orders_reviews,
    'review_creation_date < order_purchase_timestamp',
    'Отзыв раньше заказа'
)
```

## Проверка выхода за границы допустимого диапазона финансовых показателей

```python
finance_columnas = {
    'price': order_items_df,
    'freight_value': order_items_df,
    'payment_value': order_payments_df
}

def financail_features(df, column_name): 
    if len(df[df[column_name] < 0]) > 0:
        display(df[df[column_name] < 0])
        print()
    else:
        print(f'Выход за границы допустимого диапазона в колонке {column_name} не обнаружено')

for column_name, df in finance_columnas.items():
    financail_features(df, column_name)
```
